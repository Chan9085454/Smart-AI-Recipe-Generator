```python
from flask import Flask, request, render_template_string, jsonify

app = Flask(__name__)

LOCAL_RECIPES = [
    {"title": "Pasta", "ingredients": "Pasta, Tomato Sauce, Olive Oil, Garlic", "steps": "Boil pasta, add sauce, cook for 10 mins."},
    {"title": "Pizza", "ingredients": "Pizza Base, Tomato Sauce, Cheese, Toppings", "steps": "Bake at 200¬∞C for 15 mins."},
    {"title": "Biryani", "ingredients": "Rice, Chicken, Spices, Onion", "steps": "Cook chicken, layer with rice, steam for 20 mins."},
    {"title": "Sandwich", "ingredients": "Bread, Butter, Lettuce, Tomato", "steps": "Assemble ingredients between bread slices."},
    {"title": "Paneer Tikka", "ingredients": "Paneer, Yogurt, Spices, Onion", "steps": "Marinate paneer, grill for 10 mins."},
    {"title": "Dosa", "ingredients": "Rice, Urad Dal, Salt", "steps": "Ferment batter, cook on a pan until crisp."},
    {"title": "Idli", "ingredients": "Rice, Urad Dal, Salt", "steps": "Ferment batter, steam in idli maker."},
    {"title": "Upma", "ingredients": "Semolina, Water, Spices", "steps": "Roast semolina, cook with spiced water."},
    {"title": "Poha", "ingredients": "Flattened Rice, Onion, Spices", "steps": "Cook flattened rice with spices and vegetables."},
    {"title": "Chole Bhature", "ingredients": "Chickpeas, Flour, Spices", "steps": "Cook chickpeas curry, serve with fried bread."},
    {"title": "Rajma", "ingredients": "Kidney Beans, Tomato, Spices", "steps": "Cook kidney beans with tomato gravy."},
    {"title": "Dal Makhani", "ingredients": "Urad Dal, Butter, Spices", "steps": "Slow-cook dal with butter and spices."},
    {"title": "Aloo Paratha", "ingredients": "Potato, Flour, Spices", "steps": "Stuff potato mix in dough, cook on pan."},
    {"title": "Chowmein", "ingredients": "Noodles, Vegetables, Spices", "steps": "Stir-fry noodles with vegetables and sauce."},
    {"title": "Fried Rice", "ingredients": "Rice, Vegetables, Soy Sauce", "steps": "Stir-fry cooked rice with veggies and sauce."},
    {"title": "Butter Chicken", "ingredients": "Chicken, Butter, Spices", "steps": "Cook chicken in rich butter gravy."},
    {"title": "Kadhai Paneer", "ingredients": "Paneer, Onion, Capsicum", "steps": "Cook paneer with kadhai spices."},
    {"title": "Veg Pulao", "ingredients": "Rice, Vegetables, Spices", "steps": "Cook rice with vegetables and spices."},
    {"title": "Pav Bhaji", "ingredients": "Pav Bread, Mixed Veg Curry", "steps": "Cook mashed vegetables, serve with pav."},
    {"title": "Samosa", "ingredients": "Potato, Flour, Spices", "steps": "Stuff potato mix in dough, fry till crisp."},
    {"title": "Kheer", "ingredients": "Rice, Milk, Sugar", "steps": "Cook rice in milk, add sugar and dry fruits."},
    {"title": "Gulab Jamun", "ingredients": "Khoya, Sugar Syrup", "steps": "Make dough balls, fry and soak in syrup."},
    {"title": "Jalebi", "ingredients": "Flour, Sugar Syrup", "steps": "Deep-fry batter in spiral shape, dip in syrup."},
    {"title": "Rasgulla", "ingredients": "Paneer, Sugar Syrup", "steps": "Boil paneer balls in syrup."},
    {"title": "Halwa", "ingredients": "Semolina, Ghee, Sugar", "steps": "Cook semolina in ghee and sugar syrup."},
    {"title": "Dhokla", "ingredients": "Gram Flour, Spices", "steps": "Steam fermented batter, garnish with mustard seeds."},
    {"title": "Pakora", "ingredients": "Vegetables, Gram Flour, Spices", "steps": "Coat vegetables in batter, deep-fry."},
    {"title": "Bhindi Masala", "ingredients": "Ladyfinger, Onion, Spices", "steps": "Cook ladyfinger with onions and spices."},
    {"title": "Palak Paneer", "ingredients": "Spinach, Paneer, Spices", "steps": "Cook paneer in spinach gravy."},
    {"title": "Matar Paneer", "ingredients": "Peas, Paneer, Tomato Gravy", "steps": "Cook peas and paneer in tomato curry."},
    {"title": "Roti", "ingredients": "Flour, Water, Salt", "steps": "Knead dough, roll and cook on tawa."},
    {"title": "Kulcha", "ingredients": "Flour, Curd, Baking Powder", "steps": "Make dough, roll and cook in oven."},
    {"title": "Puri", "ingredients": "Flour, Salt, Oil", "steps": "Knead dough, roll and deep-fry."},
    {"title": "Gajar Halwa", "ingredients": "Carrot, Milk, Sugar", "steps": "Cook grated carrot with milk and sugar."},
    {"title": "Veg Korma", "ingredients": "Mixed Vegetables, Spices", "steps": "Cook vegetables in creamy sauce."},
    {"title": "Shahi Paneer", "ingredients": "Paneer, Cream, Spices", "steps": "Cook paneer in rich creamy sauce."}
]


ALL_DISHES = [recipe['title'] for recipe in LOCAL_RECIPES]

# Home Page HTML
HOME_PAGE = """
<!DOCTYPE html>
<html>
<head>
    <title>AI Recipe Generator</title>
    <style>
        body {font-family: Arial, sans-serif; background-color: #f0f8ff;}
        .container {background: #fff; padding: 40px; border-radius: 15px; box-shadow: 0 5px 15px rgba(0, 0, 0, 0.3); text-align: center; max-width: 600px; width: 100%; margin: 20px auto;}
        h1 {color: #FF5722;}
        p {color: #333;}
        input, button {width: 100%; padding: 10px; margin-top: 10px; border: 2px solid #FF5722; border-radius: 5px;}
        button {background-color: #FF5722; color: #fff; cursor: pointer;}
        button:hover {background-color: #E64A19;}
    </style>
</head>
<body>
    <div class="container">
        <h1>üç≤ AI Recipe Generator</h1>
        <p>Discover delicious recipes instantly!</p>
        <form action="/get_recipe" method="POST">
            <input type="text" name="food_item" placeholder="e.g., Pasta, Pizza, Biryani" required>
            <button type="submit">Get Recipe</button>
        </form>
    </div>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(HOME_PAGE)

@app.route('/get_recipe', methods=['POST'])
def get_recipe():
    food_item = request.form['food_item'].strip().lower()

    # Match exact dish name only
    found_recipe = next((recipe for recipe in LOCAL_RECIPES if recipe['title'].strip().lower() == food_item), None)

    if found_recipe:
        return jsonify(found_recipe)

    suggestions = [dish for dish in ALL_DISHES if food_item in dish.lower()]
    if suggestions:
        return jsonify({"error": "Recipe not found. Did you mean: " + ", ".join(suggestions) + "?"})

    return jsonify({"error": "Recipe not found in local data. Please try a different dish."})

if __name__ == '__main__':
    app.run(debug=True, port=69545, use_reloader=False)

```
